```python
import pandas as pnd
import numpy as np
from matplotlib import pyplot as plt

sample = pnd.read_csv("penerbangan1.csv", names=['Date', 'FlightNo', 'City1', 'City2', 'Capacity1','Capacity2'])

sample
```

```python
sample["Route"] = sample["City1"]+ "-" + sample["City2"]
sample["TotalPassengers"] = sample["Capacity1"] + sample["Capacity2"]
sample
```

```python
plt.rcParams['figure.figsize'] = (21,15)

mean1 = sum(sample.TotalPassengers)/len(sample)
print("Total passengers mean :", mean1)
x = np.array([0, 10])
y = np.array([mean1, mean1])
plt.plot(x, y, color='r', label="Rata-rata jumlah penumpang")

plt.title("Perbandingan Total Penumpang dan Rute Penerbangan")
plt.plot(sample.Route,sample.TotalPassengers,"o",label="Jumlah Penumpang")
plt.legend()
plt.xlabel("Rute Penerbangan")
plt.ylabel("Jumlah Penumpang")
plt.grid()
plt.show()
```

```python
total = 0

print("Above average passengers' flight :")
for i in range(len(sample)):
    if(sample.TotalPassengers.iloc[i] > mean1):
        total += 1
        #print("Flight :", sample.Route.iloc[i])
        
print("Total : ", total ,"flights")
print("Percentage :", (total/len(sample))*100, "%")
```

```python
total = 0

print("Below average passengers' flight :")
for i in range(len(sample)):
    if(sample.TotalPassengers.iloc[i] < mean1):
        total += 1
        #print("Flight :", sample.Route.iloc[i])
        
print("Total : ", total ,"flights")
print("Percentage :", (total/len(sample))*100, "%")
```

```python
max = 0

for i in range(len(sample)):
    temp = sample.TotalPassengers.iloc[i]
    if(temp > max):
        max = temp
        
print("Flight with most passengers:")
for i in range(len(sample)):
    if(sample.TotalPassengers.iloc[i] == max):
        print(sample.Route.iloc[i],"(",sample.FlightNo.iloc[i],")","|",sample.TotalPassengers.iloc[i])
        percent = ((sample.TotalPassengers.iloc[i])/sum(sample.TotalPassengers))*100
        print("Percentage :", percent,"%")
```

```python
min = sample.TotalPassengers.iloc[0]

for i in range(len(sample)):
    temp = sample.TotalPassengers.iloc[i]
    if(temp < min):
        min = temp
        
print("Flight with least passengers:")
for i in range(len(sample)):
    if(sample.TotalPassengers.iloc[i] == min):
        print(sample.Route.iloc[i],"(",sample.FlightNo.iloc[i],")","|",sample.TotalPassengers.iloc[i])
        percent = ((sample.TotalPassengers.iloc[i])/sum(sample.TotalPassengers))*100
        print("Percentage :", percent,"%")
```

```python
#Data departure setiap hari

def total(day):
    total = 0
    for i in range(len(sample)):
        if(day == sample.Date.iloc[i]):
            total += sample.Capacity1.iloc[i]
    
    return total

totalPass_1 = []
sum_tempDate = 0
i = 0
dayName = []

while(sum_tempDate < len(sample)):
    day = sample.Date.iloc[sum_tempDate]
    dayName.insert(i,day)
    totalPass_1.insert(i,total(day))
    sum_tempDate += sum(sample.Date == sample.Date.iloc[sum_tempDate])
    i += 1
    
print(totalPass_1)
```

```python
#Data arrival setiap hari

def total(day):
    total = 0
    for i in range(len(sample)):
        if(day == sample.Date.iloc[i]):
            total += sample.Capacity2.iloc[i]
    
    return total

totalPass_2 = []
sum_tempDate = 0
i = 0

while(sum_tempDate < len(sample)):
    day = sample.Date.iloc[sum_tempDate]
    totalPass_2.insert(i,total(day))
    sum_tempDate += sum(sample.Date == sample.Date.iloc[sum_tempDate])
    i += 1

print(totalPass_2)
```

```python
#Data total penumpang setiap hari

def total(day):
    total = 0
    for i in range(len(sample)):
        if(day == sample.Date.iloc[i]):
            total += sample.TotalPassengers.iloc[i]
    
    return total

totalPass_3 = []
sum_tempDate = 0
i = 0

while(sum_tempDate < len(sample)):
    day = sample.Date.iloc[sum_tempDate]
    totalPass_3.insert(i,total(day))
    sum_tempDate += sum(sample.Date == sample.Date.iloc[sum_tempDate])
    i += 1

print(totalPass_3)
```

```python
plt.rcParams['figure.figsize'] = (14,10)

plt.plot(dayName,totalPass_1,color='b',label='Total departed passengers')
plt.plot(dayName,totalPass_2,color='g',label='Total arrived passengers')
plt.plot(dayName,totalPass_3,color='r',label='Total passengers')
plt.title("Total passangers compared to it's date")
plt.legend()
plt.xlabel("Hari")
plt.ylabel("Jumlah Penumpang")
plt.grid()
plt.show()
```
